# Import packages

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
import pandas as pd
import numpy as np
```

# Step 1: Load data

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

# Step 2: Preprocessing

```python
# use 300 people's age, estimated ,and gender salary to train data
label_encoder = preprocessing.LabelEncoder()
encoded_Gender = label_encoder.fit_transform(data["Gender"])
data["Gender"] = encoded_Gender

X = data[['Gender', 'Age', 'EstimatedSalary']].values
Y = data[['Purchased']].values

print("X={}, Shape={}".format(X, np.shape(X)))
print("Y={}, Shape={}".format(Y, np.shape(Y)))
```

# Step 3: Train model

```python
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X,Y,test_size=0.25, random_state=0)

# Ref from https://tinyurl.com/y4pr2f4r
sc = preprocessing.StandardScaler()
Train_X = sc.fit_transform(Train_X)
Test_X = sc.transform(Test_X)

model = LR()
predY = model.fit(Train_X, Train_Y).predict(Test_X)
```

# Step 4 : Get Accuracy

```python
print("Prediction accuracy is {}".format(accuracy_score(predY, Test_Y)))
print("Prediction probability is {}".format(model.predict_proba(Test_X)))
```
